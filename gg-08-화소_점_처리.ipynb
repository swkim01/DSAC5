<h1>덧셈 연산</h1>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 이미지 읽기
!curl -L https://github.com/swkim01/DSAC5/raw/main/data/case.jpg \
    -o case.jpg
img = cv2.imread('./case.jpg')
grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(131),plt.imshow(grayimg),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(grayimg+10),plt.title('add 10')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(grayimg+50),plt.title('add 50')
plt.xticks([]), plt.yticks([])
plt.show()
```
